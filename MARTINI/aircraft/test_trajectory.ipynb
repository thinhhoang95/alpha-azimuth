```python
from trajectory import test_get_4d_trajectory_from_flight_plan

event_times, event_states, state_times, state_states = test_get_4d_trajectory_from_flight_plan()
```

```python
import numpy as np
from matplotlib import pyplot as plt

concatenated_times = np.concatenate(state_times)
concatenated_states = np.concatenate(state_states, axis=1)
```

```python
plt.plot(concatenated_times, concatenated_states[0, :], label='x')
plt.plot(concatenated_times, concatenated_states[1, :], label='y')
plt.legend()
plt.show()
```

```python
# Plot the heading angle
plt.plot(concatenated_times, np.rad2deg(concatenated_states[5, :]), label='Heading Angle')
plt.plot(concatenated_times, np.rad2deg(concatenated_states[6, :]), label='Roll Angle')
plt.legend()
plt.show()
```
